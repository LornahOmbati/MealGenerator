```python
#error handling
try:

  import re

  import requests

  request_url = "https://www.themealdb.com/api/json/v1/1/random.php"

  # Get data from the web API

  response = requests.get(request_url)

  # convert the JSON
  #data object into a python-friendly format.
  data = response.json()

  # Extract and print the meal names
  meals = data.get('meals', [])

  #loop to print all meal details required through the meals details list.
  for meal in meals:
    print("Meal id : " + meal.get('idMeal'))
    print("-------")
    print("name : " + meal.get('strMeal'))
    print("-------")
    print("Youtube link : " + meal.get('strYoutube'))
    print("-------")
    print("Article Link : " + meal.get('strSource'))
    print("-------")
    print("Instructions : ")

    # Split the paragraph into sentences
    procedure = meal.get('strInstructions')
    sentences = re.split(r'(?<=[.!?])\s+', procedure)

    #loop through the sentences and print after period.
    for sentence in sentences:
      print(sentence)

    print("-------")
    #code to interact with user.
    user = input("Do you like the meal suggestion? (y/n)?")
    if user == "y":
      print("Enjoy your meal.")
      print("-------")
    elif user == "n":
      print("Kindly run program again.")
      print("-------")
    else:
      print("Invalid response.")

except Exception as Error:
  print("Sorry, there was a problem.")
  print("-------")
  print(Error)
```
